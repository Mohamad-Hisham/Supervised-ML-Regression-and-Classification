### 1.4.2 Cost Function 

To measure how good our model is, we define a **cost function**.

For Linear Regression, we use **Mean Squared Error (MSE)**:

$$
J(w,b) = \frac{1}{2m} \sum_{i=1}^{m} \left( \hat{y}^{(i)} - y^{(i)} \right)^2
$$

 Where:

| Symbol | Meaning |
|--------|---------|
| $m$ | Number of training examples |
| $\hat{y}^{(i)} = w x^{(i)} + b$ | Predicted value |
| $y^{(i)}$ | Actual value |

---

<p align="center" style="margin:10px 0;">
  <img src="images/15.png" width="70%"/>
</p>

Dividing by $m$:

- Computes the **average error**
- Makes the cost independent of dataset size
- Keeps values numerically stable


The factor $2$ is added to simplify derivatives.

When we differentiate:

$$
\frac{d}{dw} (error)^2
$$

The $2$ cancels out nicely, making gradient descent equations cleaner (coming later).

- $\frac{1}{m}$ gives average error  
- $\frac{1}{2}$ simplifies gradient computation 

---

 We want to find parameters that minimize the cost:

$$
\min_{w,b} J(w,b)
$$

For the simpler case (only slope):

$$
\min_{w} J(w)
$$

---

<p align="center" style="margin:10px 0;">
  <img src="images/16.png" width="70%"/>
</p>

For linear regression, the cost function is:

- **Convex**
- Bowl-shaped
- Has one global minimum

That means:

Â¦ No local minima  
Â¦ Gradient descent will converge to the global minimum  

---

<p align="center" style="margin:30px ;">
  <img src="images/20.png" width="70%"/>
</p>

---

- $J(w,b)$ is **large**  
  Predictions are far from the actual values  

- $J(w,b)$ is **small**  
  Predictions are close to the actual values  

---

The optimal parameters are:

$$
w^*, \; b^*
$$

They are the values that minimize the cost function:

$$
\min_{w,b} J(w,b)
$$

cost is a measure how well our model is predicting the target price of the house.

The equation for cost with one variable is:

$$
J(w,b) = \frac{1}{2m} \sum_{i=0}^{m-1} \left( f_{w,b}(x^{(i)}) - y^{(i)} \right)^2
$$

where

$$
f_{w,b}(x^{(i)}) = wx^{(i)} + b
$$
  
- $f_{w,b}(x^{(i)})$ is our prediction for example $i$ using parameters $w,b$.  
- $(f_{w,b}(x^{(i)}) -y^{(i)})^2$ is the squared difference between the target value and the prediction.   
- These differences are summed over all the $m$ examples and divided by `2m` to produce the cost, $J(w,b)$.  
>Note that in lecture summation ranges are typically from 1 to m, while code will be from 0 to m-1.


```python
def Compute_Cost (x,y,w,b):
    """
    Computes the cost function for linear regression.
    
    Args:
      x (ndarray (m,)): Data, m examples 
      y (ndarray (m,)): target values
      w,b (scalar)    : model parameters  
    
    Returns
        total_cost (float): The cost of using w,b as the parameters for linear regression
               to fit the data points in x and y
    """
    
    m = len(x)
    cost_sum = 0
    for i in range(m):
        f_wb = w * x[i] + b
        cost = (f_wb - y[i]) ** 2
        cost_sum = cost + cost_sum
    total_cost = (1/2) * (1/m) * cost_sum
    return total_cost
```

### 1.4.3 Optimization Gradient Descent

#### What is Gradient Descent?

**Gradient Descent** is an optimization algorithm used to **minimize the cost function** of a machine learning model.

In linear regression, we want to minimize the cost function:

$$
J(w,b)
$$

which measures how wrong the model predictions are.

The algorithm works by **iteratively adjusting the parameters** \(w\) and \(b\) to reduce the cost.

#### Intuition

1. Start with some initial values of \(w\) and \(b\) (often random or zero).
2. Compute how the cost changes with respect to each parameter.
3. Move the parameters slightly in the direction that **reduces the cost**.
4. Repeat until the algorithm **converges to a minimum**.

<p align="center" style="margin:10px 0;">
  <img src="images/21.png" width="80%"/>
</p>

#### Gradient Descent Update Rule

At each iteration, we update the parameters using:

$$
w = w - \alpha \frac{\partial J(w,b)}{\partial w}
$$

$$
b = b - \alpha \frac{\partial J(w,b)}{\partial b}
$$

Where:

- $J(w,b)$ **Cost function**
- $\alpha$ **Learning rate** (step size , small positive number)
- $\frac{\partial J}{\partial w}$ and $\frac{\partial J}{\partial b}$ **Partial derivatives (gradients)**

These derivatives tell us **how sensitive the cost function is to each parameter**.

---

#### Why Do We Subtract the Gradient?

The gradient points in the direction of **steepest increase** of the function.

To **minimize** the function, we move in the **opposite direction**, which is why we subtract the gradient.

> **Key Idea:**  
> Move in the direction of the **steepest decrease** of the cost.

<p align="center" style="margin:10px 0;">
  <img src="images/23.png" width="75%"/>
</p>

---

#### Simultaneous Parameter Update

Both parameters must be updated **simultaneously**.

Correct implementation:

$$
\text{temp\_w} = w - \alpha \frac{\partial J(w,b)}{\partial w}
$$

$$
\text{temp\_b} = b - \alpha \frac{\partial J(w,b)}{\partial b}
$$

Then update:
w = temp_w and b = temp_b

If we update \(w\) first and then use the new value when updating \(b\), the algorithm becomes incorrect.

---

<p align="center" style="margin:10px 0;">
  <img src="images/22.png" width="75%"/>
</p>

---

**Geometric Interpretation**

Imagine standing on a **mountain in fog**.

You want to reach the **lowest point in the valley**, but you can only feel the slope around you.

So you repeatedly:

1. Check the steepest downward direction
2. Take a step in that direction
3. Repeat until you reach the bottom

This is exactly what **Gradient Descent** does with the cost function.

---

**gradient descent from the cost function**

To apply gradient descent we compute the **partial derivatives** of the cost function. Gradient with respect to $w$: $$ \frac{\partial J(w,b)}{\partial w} = \frac{1}{m} \sum_{i=1}^{m} \left(f_{w,b}(x^{(i)}) - y^{(i)}\right) x^{(i)} $$
Gradient with respect to $b$: $$ \frac{\partial J(w,b)}{\partial b} = \frac{1}{m} \sum_{i=1}^{m} \left(f_{w,b}(x^{(i)}) - y^{(i)}\right) $$
<p align ="center" style ="margin:10px 0;"> <img src = "images/27.png" width = "75%"/> </p>

#### Gradient Descent Algorithm


$$
w = w - \alpha
\left(
\frac{1}{m}
\sum_{i=1}^{m}
\left(f_{w,b}(x^{(i)}) - y^{(i)}\right) x^{(i)}
\right)
$$

$$
b = b - \alpha
\left(
\frac{1}{m}
\sum_{i=1}^{m}
\left(f_{w,b}(x^{(i)}) - y^{(i)}\right)
\right)
$$

Both parameters **must be updated simultaneously**.

#### Batch Gradient Descent

In **Batch Gradient Descent**, each step of gradient descent uses **all the training examples** in the dataset.

If we have $m$ training examples:

$$
(x^{(1)}, y^{(1)}), (x^{(2)}, y^{(2)}), \dots, (x^{(m)}, y^{(m)})
$$

the algorithm computes the gradient using **all $m$ examples** before updating the parameters.

Batch Gradient Descent computes the **average error across the entire dataset** before updating the parameters.

Because it uses all training examples, the gradient direction is **very accurate**, but each step can be **computationally expensive for large datasets**.


At every iteration:

1. Compute the prediction error for **all $m$ examples**
2. Average the gradients
3. Update $w$ and $b$

---

**Example Analogy**

Imagine grading **47 exams** and deciding the class average.

You first read **all 47 exams**, compute the average score, and only then decide what adjustment to make.

This is exactly how **Batch Gradient Descent** works.

```python
import math, copy
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('./deeplearning.mplstyle')
from lab_utils_uni import plt_house_x, plt_contour_wgrad, plt_divergence, plt_gradients
```

```python
# Load our data set
x_train = np.array([1.0, 2.0])   #features
y_train = np.array([300.0, 500.0])   #target value
```

```python
def compute_cost(x,y,w,b):
    m=x.shape[0]
    cost=0
    for i in range(m):
        f_wb=w*x[i] + b
        cost =cost + (f_wb - y[i])**2
        total_cost = 1/(2*m) * cost
    return total_cost
```

```python
def compute_gradient (x , y , w , b):
    """
    Computes the gradient for linear regression 
    Args:
      x (ndarray (m,)): Data, m examples 
      y (ndarray (m,)): target values
      w,b (scalar)    : model parameters  
    Returns
      dj_dw (scalar): The gradient of the cost w.r.t. the parameters w
      dj_db (scalar): The gradient of the cost w.r.t. the parameter b     
      """
    m= x.shape[0]
    dj_db = 0
    dj_dw = 0
    for i in range(m):
        f_wb= w* x[i] +b
        dj_dw += (1/m) * (f_wb - y[i]) * x[i]
        dj_db += (1/m) * (f_wb - y[i])
    return dj_dw , dj_db
```

```python
plt_gradients(x_train,y_train, compute_cost, compute_gradient)
plt.show()
```

```python
def gradient_descent (x,y,w_in,b_in,alpha,num_iterations,cost_function,gredient_function):
      # An array to store cost J and w's at each iteration primarily for graphing later
    J_history = []
    p_history = []
    w=copy.deepcopy(w_in)
    b=copy.deepcopy(b_in)
    for i in range(num_iterations):
        dj_dw , dj_db = gredient_function(x,y,w,b)
        w= w - alpha * dj_dw
        b= b - alpha * dj_db
         #prevent resource exhaustion 
        if i<100000:      
            J_history.append( cost_function(x, y, w , b))
            p_history.append([w,b])

        if i% math.ceil(num_iterations/10) == 0:
            print(f"Iteration {i:4}: Cost {J_history[-1]:0.2e} ",
                  f"dj_dw: {dj_dw: 0.3e}, dj_db: {dj_db: 0.3e}  ",
                  f"w: {w: 0.3e}, b:{b: 0.5e}")
            
    return w, b, J_history, p_history
```

```python
w_init = 0
b_init = 0
iterations = 10000
tmp_alpha = 1.0e-2
w_final, b_final, J_hist, p_hist = gradient_descent(x_train ,y_train, w_init, b_init, tmp_alpha, 
                                                    iterations, compute_cost, compute_gradient)
print(f"(w,b) found by gradient descent: ({w_final:8.4f},{b_final:8.4f})")
```

# 2 Regression with multiple input variables

## 2.1 Model Explaination


###  Feature Notation

A single training example is written as:

$$
\mathbf{x}^{(i)} = \begin{bmatrix} x_1^{(i)} & x_2^{(i)} & x_3^{(i)} & \cdots & x_n^{(i)} \end{bmatrix}
$$

Where:

$$
x_j = \text{the } j^{\text{th}} \text{ feature}
$$

$$
n = \text{number of features}
$$

$$
x_j^{(i)} = \text{value of feature } j \text{ in training example } i
$$

Example:

$$
\mathbf{x}^{(2)} = \begin{bmatrix} 1416 & 3 & 2 & 40 \end{bmatrix}
$$

$$
x_3^{(2)} = 2
$$

This means the third feature of the second training example equals 2.

<div align="center">
    <img src="images/Multiple input features.png" width="650">
</div>

---

### Model Formula

For a single input variable, the model was:

$$
f_{w,b}(x) = wx + b
$$

For multiple input variables, the model becomes:

$$
f_{\mathbf{w},b}(\mathbf{x}) = w_1x_1 + w_2x_2 + w_3x_3 + \cdots + w_nx_n + b
$$

In vector form:

$$
f_{\mathbf{w},b}(\mathbf{x}) = \mathbf{w} \cdot \mathbf{x} + b
$$

---

### Parameter Vectors

$$
\mathbf{w} = \begin{bmatrix} w_1 & w_2 & w_3 & \cdots & w_n \end{bmatrix}
$$

$$
\mathbf{x} = \begin{bmatrix} x_1 & x_2 & x_3 & \cdots & x_n \end{bmatrix}
$$

$$
b \in \mathbb{R}
$$

Here:
- $\mathbf{w}$ is the weight vector
- $\mathbf{x}$ is the feature vector
- $\mathbf{w} \cdot \mathbf{x}$ is the dot product
- $b$ is the bias term

---

### House Price Example

Suppose the features are:

$$
x_1 = \text{size}, \qquad x_2 = \text{bedrooms}, \qquad x_3 = \text{floors}, \qquad x_4 = \text{age}
$$

Then one possible model is:

$$
f_{\mathbf{w},b}(\mathbf{x}) = 0.1x_1 + 4x_2 + 10x_3 - 2x_4 + 80
$$

---

### Important Note

$$
\text{Multiple linear regression} \; \neq \; \text{multivariate regression}
$$

It is called **multiple linear regression** because we have **multiple input features** but still **one output**.

## 2.2 Vactorization

Vectorization means performing operations on whole arrays at once instead of looping through each element one by one. In machine learning, this makes code shorter, cleaner, and much faster because libraries like NumPy can execute many operations efficiently in parallel.

### Parameters and Features as Vectors

In multiple linear regression, the parameters and input features can be written as vectors:

$$
\mathbf{w} = [w_1, w_2, \dots, w_n]
$$

$$
b \in \mathbb{R}
$$

$$
\mathbf{x} = [x_1, x_2, \dots, x_n]
$$

For example, when \(n = 3\):

$$
\mathbf{w} = [1.0, 2.5, -3.3], \quad b = 4, \quad \mathbf{x} = [10, 20, 30]
$$

### Without Vectorization

The model prediction can be computed by summing one term at a time:

$$
f_{\mathbf{w},b}(\mathbf{x}) = \sum_{j=1}^{n} w_j x_j + b
$$

For \(n = 3\), this becomes:

$$
f_{\mathbf{w},b}(\mathbf{x}) = w_1x_1 + w_2x_2 + w_3x_3 + b
$$

Using the example values:

$$
f_{\mathbf{w},b}(\mathbf{x}) = (1.0)(10) + (2.5)(20) + (-3.3)(30) + 4
$$

$$
f_{\mathbf{w},b}(\mathbf{x}) = 10 + 50 - 99 + 4 = -35
$$

This approach usually needs a loop that processes each feature separately.

### With Vectorization

Instead of computing each term one by one, we use the dot product:

$$
f_{\mathbf{w},b}(\mathbf{x}) = \mathbf{w} \cdot \mathbf{x} + b
$$

$$
\mathbf{w} \cdot \mathbf{x} = \sum_{j=1}^{n} w_j x_j
$$

So the same example can be written compactly as:

$$
f_{\mathbf{w},b}(\mathbf{x}) = [1.0, 2.5, -3.3] \cdot [10, 20, 30] + 4 = -35
$$

In NumPy, this is done with:

$$
f = \operatorname{np.dot}(\mathbf{w}, \mathbf{x}) + b
$$

    <img src="images/vactorization (1).png" width="650">
</div>

### Why Vectorization Is Better

Without vectorization, the computer handles each multiplication and addition step by step:

$$
f + w_1x_1 \rightarrow f
$$

$$
f + w_2x_2 \rightarrow f
$$

$$
\vdots
$$

$$
f + w_nx_n \rightarrow f
$$

<div align="center">
    <img src="images/vactorization (2).png" width="650">
</div>

With vectorization, many of these operations are handled together internally, which is much more efficient, especially when $n$ is very large.

### Vectorized Gradient Descent Update

Vectorization is also useful when updating parameters during gradient descent. If:

$$
\mathbf{w} = [w_1, w_2, \dots, w_n]
$$

$$
\mathbf{d} = [d_1, d_2, \dots, d_n]
$$

where $\mathbf{d}$ contains the derivatives of the cost function with respect to each parameter, then the element-by-element update is:

$$
w_j = w_j - \alpha d_j, \quad j = 1,2,\dots,n
$$

The vectorized form is:

$$
\mathbf{w} = \mathbf{w} - \alpha \mathbf{d}
$$

For example, if $\alpha = 0.1$, then:

$$
\mathbf{w} = \mathbf{w} - 0.1\mathbf{d}
$$

This updates all parameters at the same time and avoids writing a separate loop for each component.

<div align="center">
    <img src="images/vactorization (3).png" width="650">
</div>

### Summary

Vectorization replaces explicit loops with whole-vector operations. In linear regression, it lets us compute predictions using the dot product and update parameters using one compact gradient descent equation. This makes the implementation simpler and significantly faster, especially for models with many features.

## 2.3 Gradient descent for multiple linear regression

In multiple linear regression, gradient descent updates all parameters by moving them in the direction that reduces the cost function. The main difference from the one-feature case is that we now have one weight for each feature, so each weight has its own partial derivative.

### Review: One-Feature Gradient Descent

For one feature, the updates are:

$$
w = w - \alpha \frac{1}{m} \sum_{i=1}^{m} \left(f_{w,b}(x^{(i)}) - y^{(i)}\right)x^{(i)}
$$

$$
b = b - \alpha \frac{1}{m} \sum_{i=1}^{m} \left(f_{w,b}(x^{(i)}) - y^{(i)}\right)
$$

Both parameters must be updated simultaneously after computing the derivatives.

### Gradient Descent for Multiple Features

When there are multiple features, we have parameters:

$$
\mathbf{w} = [w_1, w_2, \dots, w_n]
$$

The prediction function is:

$$
f_{\mathbf{w},b}(\mathbf{x}^{(i)}) = \mathbf{w} \cdot \mathbf{x}^{(i)} + b
$$

Each parameter is updated using its own derivative:

$$
w_j = w_j - \alpha \frac{1}{m} \sum_{i=1}^{m} \left(f_{\mathbf{w},b}(\mathbf{x}^{(i)}) - y^{(i)}\right)x_j^{(i)}, \quad j = 1,2,\dots,n
$$

The bias update remains:

$$
b = b - \alpha \frac{1}{m} \sum_{i=1}^{m} \left(f_{\mathbf{w},b}(\mathbf{x}^{(i)}) - y^{(i)}\right)
$$

These updates are also done simultaneously, meaning we first compute all new values, then assign them together.

### Meaning of the Weight Update

For each feature index:

$$
j = 1,2,\dots,n
$$

the term:

$$
\frac{1}{m} \sum_{i=1}^{m} \left(f_{\mathbf{w},b}(\mathbf{x}^{(i)}) - y^{(i)}\right)x_j^{(i)}
$$

is the partial derivative of the cost function with respect to the parameter:

$$
\frac{\partial}{\partial w_j} J(\mathbf{w}, b)
$$

So gradient descent updates each weight by subtracting the learning rate times its derivative.

### Alternative to Gradient Descent: Normal Equation

For linear regression, there is another method called the normal equation. It solves directly for the parameters without iterative updates.

Instead of repeatedly applying:

$$
w_j = w_j - \alpha \frac{1}{m} \sum_{i=1}^{m} \left(f_{\mathbf{w},b}(\mathbf{x}^{(i)}) - y^{(i)}\right)x_j^{(i)}
$$

the normal equation computes the solution in one direct procedure.

### Important Notes About the Normal Equation

- It is used only for linear regression.
- It solves for the parameters without iterations.
- It does not generalize well to many other learning algorithms.
- It becomes slow when the number of features is very large.

In practice, machine learning libraries may implement the normal equation method for linear regression, but gradient descent is more general and is the more important method to understand.

```python
def predict(x, w, b): 
    """
    single predict using linear regression
    Args:
      x (ndarray): Shape (n,) example with multiple features
      w (ndarray): Shape (n,) model parameters   
      b (scalar):             model parameter 
      
    Returns:
      p (scalar):  prediction
    """
    p = np.dot(x, w) + b     
    return p    
```

The equation for the cost function with multiple variables $J(\mathbf{w}, b)$ is:

$$
J(\mathbf{w}, b) = \frac{1}{2m} \sum_{i=0}^{m-1}
\left(f_{\mathbf{w},b}(\mathbf{x}^{(i)}) - y^{(i)}\right)^2
$$

where:

$$
f_{\mathbf{w},b}(\mathbf{x}^{(i)}) = \mathbf{w} \cdot \mathbf{x}^{(i)} + b
$$


```python
def compute_cost(X, y, w, b): 
    """
    compute cost
    Args:
      X (ndarray (m,n)): Data, m examples with n features
      y (ndarray (m,)) : target values
      w (ndarray (n,)) : model parameters  
      b (scalar)       : model parameter
      
    Returns:
      cost (scalar): cost
    """
    m = X.shape[0]
    cost = 0.0
    for i in range(m):                                
        f_wb_i = np.dot(X[i], w) + b           #(n,)(n,) = scalar (see np.dot)
        cost = cost + (f_wb_i - y[i])**2       #scalar
    cost = cost / (2 * m)                      #scalar    
    return cost
```

```python
cost = compute_cost(X_train, y_train, w_init, b_init)
print(f'Cost at optimal w : {cost}')
```

Gradient descent for multiple variables:

$$\begin{align*} \text{repeat}&\text{ until convergence:} \; \lbrace \newline\;
& w_j = w_j -  \alpha \frac{\partial J(\mathbf{w},b)}{\partial w_j}   \; & \text{for j = 0..n-1}\newline
&b\ \ = b -  \alpha \frac{\partial J(\mathbf{w},b)}{\partial b}  \newline \rbrace
\end{align*}$$

where, n is the number of features, parameters $w_j$,  $b$, are updated simultaneously and where  

$$
\begin{align}
\frac{\partial J(\mathbf{w},b)}{\partial w_j}  &= \frac{1}{m} \sum\limits_{i = 0}^{m-1} (f_{\mathbf{w},b}(\mathbf{x}^{(i)}) - y^{(i)})x_{j}^{(i)}   \\
\frac{\partial J(\mathbf{w},b)}{\partial b}  &= \frac{1}{m} \sum\limits_{i = 0}^{m-1} (f_{\mathbf{w},b}(\mathbf{x}^{(i)}) - y^{(i)}) 
\end{align}
$$
* m is the number of training examples in the data set

```python
def compute_gradient(X, y, w, b): 
    """
    Computes the gradient for linear regression 
    Args:
      X (ndarray (m,n)): Data, m examples with n features
      y (ndarray (m,)) : target values
      w (ndarray (n,)) : model parameters  
      b (scalar)       : model parameter
      
    Returns:
      dj_dw (ndarray (n,)): The gradient of the cost w.r.t. the parameters w. 
      dj_db (scalar):       The gradient of the cost w.r.t. the parameter b. 
    """
    m,n = X.shape           #(number of examples, number of features)
    dj_dw = np.zeros((n,))
    dj_db = 0.

    for i in range(m):                             
        err = (np.dot(X[i], w) + b) - y[i]   
        for j in range(n):                         
            dj_dw[j] = dj_dw[j] + err * X[i, j]    
        dj_db = dj_db + err                        
    dj_dw = dj_dw / m                                
    dj_db = dj_db / m                                
        
    return dj_db, dj_dw
```

```python
#Compute and display gradient 
tmp_dj_db, tmp_dj_dw = compute_gradient(X_train, y_train, w_init, b_init)
print(f'dj_db at initial w,b: {tmp_dj_db}')
print(f'dj_dw at initial w,b: \n {tmp_dj_dw}')
```

```python
def gradient_descent(X, y, w_in, b_in, cost_function, gradient_function, alpha, num_iters): 
    """
    Performs batch gradient descent to learn w and b. Updates w and b by taking 
    num_iters gradient steps with learning rate alpha
    
    Args:
      X (ndarray (m,n))   : Data, m examples with n features
      y (ndarray (m,))    : target values
      w_in (ndarray (n,)) : initial model parameters  
      b_in (scalar)       : initial model parameter
      cost_function       : function to compute cost
      gradient_function   : function to compute the gradient
      alpha (float)       : Learning rate
      num_iters (int)     : number of iterations to run gradient descent
      
    Returns:
      w (ndarray (n,)) : Updated values of parameters 
      b (scalar)       : Updated value of parameter 
      """
    
    # An array to store cost J and w's at each iteration primarily for graphing later
    J_history = []
    w = copy.deepcopy(w_in)  #avoid modifying global w within function
    b = b_in
    
    for i in range(num_iters):

        # Calculate the gradient and update the parameters
        dj_db,dj_dw = gradient_function(X, y, w, b)   ##None

        # Update Parameters using w, b, alpha and gradient
        w = w - alpha * dj_dw               ##None
        b = b - alpha * dj_db               ##None
      
        # Save cost J at each iteration
        if i<100000:      # prevent resource exhaustion 
            J_history.append( cost_function(X, y, w, b))

        # Print cost every at intervals 10 times or as many iterations if < 10
        if i% math.ceil(num_iters / 10) == 0:
            print(f"Iteration {i:4d}: Cost {J_history[-1]:8.2f}   ")
        
    return w, b, J_history #return final w,b and J history for graphing
```

```python
# initialize parameters
initial_w = np.zeros_like(w_init)
initial_b = 0.
# some gradient descent settings
iterations = 1000
alpha = 5.0e-7
# run gradient descent 
w_final, b_final, J_hist = gradient_descent(X_train, y_train, initial_w, initial_b,
                                                    compute_cost, compute_gradient, 
                                                    alpha, iterations)
print(f"b,w found by gradient descent: {b_final:0.2f},{w_final} ")
m,_ = X_train.shape
for i in range(m):
    print(f"prediction: {np.dot(X_train[i], w_final) + b_final:0.2f}, target value: {y_train[i]}")
```

## 2.4 Feature Scaling and Engineering




Feature scaling rescales input features so they have comparable numeric ranges. This is especially important when one feature is measured on a large scale, such as house size in square feet, while another feature is measured on a small scale, such as number of bedrooms.

When feature ranges are very different, the parameters usually compensate in the opposite direction: large-valued features tend to need smaller weights, and small-valued features tend to need larger weights. This can make the cost-function contours long and narrow, causing gradient descent to zig-zag and converge slowly. After scaling, the contours are more balanced, so gradient descent can move more directly toward the minimum.

### Simple Feature Scaling

Simple feature scaling divides each feature by its maximum value. This usually maps values into a range close to 0 to 1.

$$
x_{j,\mathrm{scaled}} = \frac{x_j}{x_{j,\max}}
$$

Example for house size:

$$
300 \leq x_1 \leq 2000
$$

$$
x_{1,\mathrm{scaled}} = \frac{x_1}{2000}
$$

$$
0.15 \leq x_{1,\mathrm{scaled}} \leq 1
$$

Example for number of bedrooms:

$$
0 \leq x_2 \leq 5
$$

$$
x_{2,\mathrm{scaled}} = \frac{x_2}{5}
$$

$$
0 \leq x_{2,\mathrm{scaled}} \leq 1
$$

### Mean Normalization

Mean normalization shifts each feature so its values are centered around 0, then divides by the feature range. This often gives values roughly between -1 and 1.

$$
x_{j,\mathrm{norm}} = \frac{x_j - \mu_j}{x_{j,\max} - x_{j,\min}}
$$

The mean of feature j is:

$$
\mu_j = \frac{1}{m}\sum_{i=1}^{m} x_j^{(i)}
$$

Example for house size, using mean 600:

$$
\mu_1 = 600
$$

$$
x_{1,\mathrm{norm}} = \frac{x_1 - \mu_1}{2000 - 300}
$$

$$
-0.18 \leq x_{1,\mathrm{norm}} \leq 0.82
$$

Example for number of bedrooms, using mean 2.3:

$$
\mu_2 = 2.3
$$

$$
x_{2,\mathrm{norm}} = \frac{x_2 - \mu_2}{5 - 0}
$$

$$
-0.46 \leq x_{2,\mathrm{norm}} \leq 0.54
$$

### Z-Score Normalization

Z-score normalization subtracts the mean and divides by the standard deviation. This is one of the most common scaling methods because it makes each feature have mean 0 and standard deviation 1.

$$
x_{j,\mathrm{z}} = \frac{x_j - \mu_j}{\sigma_j}
$$

If using the sample standard deviation, the sample variance is:

$$
s_j^2 = \frac{\sum_{i=1}^{m}\left(x_j^{(i)} - \mu_j\right)^2}{m - 1}
$$

The sample standard deviation is:

$$
s_j = \sqrt{s_j^2}
$$

Example for house size, using mean 600 and standard deviation 450:

$$
\mu_1 = 600, \quad \sigma_1 = 450
$$

$$
x_{1,\mathrm{z}} = \frac{x_1 - \mu_1}{\sigma_1}
$$

$$
-0.67 \leq x_{1,\mathrm{z}} \leq 3.1
$$

Example for number of bedrooms, using mean 2.3 and standard deviation 1.4:

$$
\mu_2 = 2.3, \quad \sigma_2 = 1.4
$$

$$
x_{2,\mathrm{z}} = \frac{x_2 - \mu_2}{\sigma_2}
$$

$$
-1.6 \leq x_{2,\mathrm{z}} \leq 1.9
$$

### Choosing Good Feature Ranges

A useful goal is to make each feature take values in a similar range. The ideal target is often around:

$$
-1 \leq x_j \leq 1
$$

These ranges are usually still acceptable:

$$
-3 \leq x_j \leq 3
$$

$$
-0.3 \leq x_j \leq 0.3
$$

Examples that are usually okay without rescaling:

$$
0 \leq x_1 \leq 3
$$

$$
-2 \leq x_2 \leq 0.5
$$

Examples that should usually be rescaled:

$$
-100 \leq x_3 \leq 100
$$

$$
-0.001 \leq x_4 \leq 0.001
$$

$$
98.6 \leq x_5 \leq 105
$$

### Feature engineering and Polynomial regression

Feature engineering means creating new useful features from the original features. The goal is to give the model better information so it can fit the data more accurately.

For example, suppose a house lot has frontage and depth:

$$
x_1 = \text{frontage}
$$

$$
x_2 = \text{depth}
$$

A useful new feature could be the area:

$$
x_3 = x_1 x_2
$$

The model can then use the original features and the new feature together:

$$
f_{\mathbf{w}, b}(\mathbf{x}) = w_1x_1 + w_2x_2 + w_3x_3 + b
$$

Polynomial regression is another type of feature engineering. Instead of only using a feature such as house size, we can add powers of that feature to help the model fit curved patterns.

$$
f_{\mathbf{w}, b}(x) = w_1x + w_2x^2 + b
$$

We can also use a higher-degree polynomial:

$$
f_{\mathbf{w}, b}(x) = w_1x + w_2x^2 + w_3x^3 + b
$$

Sometimes other transformations can work better than powers, such as using the square root of a feature:

$$
f_{\mathbf{w}, b}(x) = w_1x + w_2\sqrt{x} + b
$$

When using polynomial features, feature scaling becomes very important because the ranges can become very different. For example, if size ranges from 1 to 1000, then the squared and cubed features become much larger:

$$
1 \leq x \leq 10^3
$$

$$
1 \leq x^2 \leq 10^6
$$

$$
1 \leq x^3 \leq 10^9
$$

Key idea: feature engineering lets us design better inputs for the model, and polynomial regression lets a linear regression model fit curved relationships.


## 2.5 Gradient descent in practice

Gradient descent in practice is about three things: Feature scaling , checking that the cost is moving in the right direction, and choosing a learning rate $\alpha$ that is not too large or too small.

The goal is to find the parameter vector $\mathbf{w}$ and bias $b$ that minimize the cost function:

$$
\min_{\mathbf{w}, b} J(\mathbf{w}, b)
$$

For multiple linear regression, the model prediction is:

$$
f_{\mathbf{w}, b}(\mathbf{x}^{(i)}) = \mathbf{w} \cdot \mathbf{x}^{(i)} + b
$$

The cost function is:

$$
J(\mathbf{w}, b) = \frac{1}{2m}\sum_{i=1}^{m}\left(f_{\mathbf{w}, b}(\mathbf{x}^{(i)}) - y^{(i)}\right)^2
$$

####  Check the Learning Curve

A learning curve plots $J(\mathbf{w}, b)$ against the number of iterations. It shows whether gradient descent is reducing the cost over time.

If gradient descent is working correctly, $J(\mathbf{w}, b)$ should decrease after every iteration:

$$
J(\mathbf{w}^{(k+1)}, b^{(k+1)}) \leq J(\mathbf{w}^{(k)}, b^{(k)})
$$

At the beginning, the cost may decrease quickly. Later, as the parameters get closer to the minimum, the curve becomes flatter. A nearly flat learning curve usually means gradient descent has converged.

####  Automatic Convergence Test

A convergence test checks whether the cost is still decreasing by a meaningful amount. Let $\epsilon$ be a small number, for example:

$$
\epsilon = 10^{-3} = 0.001
$$

If the cost decreases by less than or equal to $\epsilon$ in one iteration, the algorithm can be declared converged:

$$
J(\mathbf{w}^{(k)}, b^{(k)}) - J(\mathbf{w}^{(k+1)}, b^{(k+1)}) \leq \epsilon
$$

This means the parameters $\mathbf{w}$ and $b$ are probably close to a minimum because each new update changes the cost only a tiny amount.

  <img
    src="images/gradient descent in practice (1).png"
    style="max-width:760px; width:90%; height:auto;"
  />
  <figcaption style="font-size:0.9em; color:#666; margin-top:8px;">
    gradient descent and cost function effect
  </figcaption>
</figure>

####  Use the Correct Update Direction

Gradient descent must subtract the derivative. The minus sign is important because it moves the parameters downhill toward a smaller cost.

$$
w_j := w_j - \alpha \frac{\partial}{\partial w_j}J(\mathbf{w}, b)
$$

$$
b := b - \alpha \frac{\partial}{\partial b}J(\mathbf{w}, b)
$$

Using a plus sign instead would move the parameters in the wrong direction:

$$
w_j := w_j + \alpha \frac{\partial}{\partial w_j}J(\mathbf{w}, b)
$$

This can make $J(\mathbf{w}, b)$ increase instead of decrease.

For multiple linear regression, the derivatives are:

$$
\frac{\partial}{\partial w_j}J(\mathbf{w}, b) = \frac{1}{m}\sum_{i=1}^{m}\left(f_{\mathbf{w}, b}(\mathbf{x}^{(i)}) - y^{(i)}\right)x_j^{(i)}
$$

$$
\frac{\partial}{\partial b}J(\mathbf{w}, b) = \frac{1}{m}\sum_{i=1}^{m}\left(f_{\mathbf{w}, b}(\mathbf{x}^{(i)}) - y^{(i)}\right)
$$

####  Diagnose the Learning Rate

The learning rate $\alpha$ controls the size of each update step.

If $\alpha$ is too large, gradient descent can overshoot the minimum. The learning curve may oscillate, rise, or diverge. A clear warning sign is:

$$
J(\mathbf{w}^{(k+1)}, b^{(k+1)}) > J(\mathbf{w}^{(k)}, b^{(k)})
$$

If $\alpha$ is too small, the cost usually still decreases, but the algorithm may need many more iterations to converge.

With a small enough $\alpha$, the cost should decrease monotonically:

$$
J(\mathbf{w}^{(0)}, b^{(0)}) \geq J(\mathbf{w}^{(1)}, b^{(1)}) \geq J(\mathbf{w}^{(2)}, b^{(2)}) \geq \cdots
$$

  <img

    style="max-width:760px; width:90%; height:auto;"
  />
  <figcaption style="font-size:0.9em; color:#666; margin-top:8px;">
    Alpha effect on the cost function and gradient descent 
  </figcaption>
</figure>

####  Choose Practical Values for $\alpha$

A practical strategy is to try learning rates that increase by about 3 times each step:

$$
\alpha \in \{0.001, 0.003, 0.01, 0.03, 0.1, 0.3, 1\}
$$

Then compare the learning curves. Choose the largest $\alpha$ that still makes $J(\mathbf{w}, b)$ decrease smoothly.

A good $\alpha$ makes the cost decrease quickly and then flatten near convergence. A too-small $\alpha$ decreases the cost very slowly. A too-large $\alpha$ makes the cost jump around or increase.

#### Key Takeaway

To use gradient descent well in practice, always monitor the learning curve. If the cost increases or oscillates, use a smaller $\alpha$ or check the update sign. If the cost decreases too slowly, try a larger $\alpha$.

```python
def zscore_normalize_features(X):
    """
    computes  X, zcore normalized by column
    
    Args:
      X (ndarray (m,n))     : input data, m examples, n features
      
    Returns:
      X_norm (ndarray (m,n)): input normalized by column
      mu (ndarray (n,))     : mean of each feature
      sigma (ndarray (n,))  : standard deviation of each feature
    """
    # find the mean of each column/feature
    mu     = np.mean(X, axis=0)                 # mu will have shape (n,)
    # find the standard deviation of each column/feature
    sigma  = np.std(X, axis=0)                  # sigma will have shape (n,)
    # element-wise, subtract mu for that column from each example, divide by std for that column
    X_norm = (X - mu) / sigma      

    return (X_norm, mu, sigma)
 
```

# 3 Classification

## 3.1 Classification with logistic regression

Classification predicts a discrete class or category, rather than a continuous number. In binary classification, the target can only take two possible values:

$$
y \in \{0, 1\}
$$

The value 0 is usually used for the negative class, such as no, false, absence, or not malignant:

$$
y = 0
$$

The value 1 is usually used for the positive class, such as yes, true, presence, or malignant:

$$
y = 1
$$

Examples of classification questions include whether an email is spam, whether a transaction is fraudulent, or whether a tumor is malignant.

A simple idea is to use the linear regression model and apply a threshold:

$$
f_{w,b}(x) = wx + b
$$

Using a threshold of 0.5, the prediction rule is:

$$
\hat{y} = 0 \quad \text{if} \quad f_{w,b}(x) < 0.5
$$

$$
\hat{y} = 1 \quad \text{if} \quad f_{w,b}(x) \geq 0.5
$$

However, a straight-line model can be a poor choice for classification. New examples far from the main data can change the fitted line and move the decision boundary, which can cause more misclassifications.

Logistic regression is designed for classification because it maps the model output to a value between 0 and 1:

$$
f_{w,b}(x) = g(wx + b)
$$

$$
g(z) = \frac{1}{1 + e^{-z}}
$$

$$
0 \leq f_{w,b}(x) \leq 1
$$

This makes the output useful for deciding between the two classes:

$$
\hat{y} = 0 \quad \text{if} \quad f_{w,b}(x) < 0.5
$$

$$
\hat{y} = 1 \quad \text{if} \quad f_{w,b}(x) \geq 0.5
$$

###  Logistic Regression

For classification, we want the model output to behave like a probability. That means the prediction should stay between zero and one.

$$
0 < f_{\vec{w},b}(\vec{x}) < 1
$$

Logistic regression does this by first computing a linear value:

$$
z = \vec{w} \cdot \vec{x} + b
$$

Then it passes that value through the sigmoid function, also called the logistic function:

$$
g(z) = \frac{1}{1 + e^{-z}}
$$

The sigmoid function always outputs a value between zero and one:

$$
0 < g(z) < 1
$$

<figure style="text-align:center; margin:24px 0;">
  <img
    src="images/logistic regression (1).png"
    style="max-width:760px; width:90%; height:auto;"
  />
  <figcaption style="font-size:0.9em; color:#666; margin-top:8px;">
    Sigmoid Function
  </figcaption>
</figure>

So the logistic regression model is:

$$
f_{\vec{w},b}(\vec{x}) = g(\vec{w} \cdot \vec{x} + b)
$$

$$
f_{\vec{w},b}(\vec{x}) = \frac{1}{1 + e^{-(\vec{w} \cdot \vec{x} + b)}}
$$

When the linear value is very large and positive, the sigmoid output gets close to one:

$$
z \gg 0 \quad \Rightarrow \quad g(z) \approx 1
$$

When the linear value is very large and negative, the sigmoid output gets close to zero:

$$
z \ll 0 \quad \Rightarrow \quad g(z) \approx 0
$$

At zero, the sigmoid output is exactly one half:

$$
z = 0 \quad \Rightarrow \quad g(z) = \frac{1}{1 + e^0} = \frac{1}{2} = 0.5
$$

The output of logistic regression is interpreted as the probability that the class is one, given the input features and the model parameters:

$$
f_{\vec{w},b}(\vec{x}) = P(y = 1 \mid \vec{x}; \vec{w}, b)
$$

Because this is binary classification, the probabilities of the two classes must add up to one:

$$
P(y = 0) + P(y = 1) = 1
$$

For example, if the input is tumor size and the model outputs the following value:

$$
f_{\vec{w},b}(\vec{x}) = 0.7
$$

Then the model is saying there is a seventy percent chance that the tumor belongs to the positive class:

$$
P(y = 1 \mid \vec{x}; \vec{w}, b) = 0.7
$$

If the positive class means malignant, then:

$$
P(\text{malignant} \mid \vec{x}; \vec{w}, b) = 0.7
$$

### Decision Boundaries

Logistic regression is used for binary classification, where the target value is one of two classes. The model first computes a linear value:

$$
z = \vec{w} \cdot \vec{x} + b
$$

Then it passes this value through the sigmoid function:

$$
g(z) = \frac{1}{1 + e^{-z}}
$$

So the logistic regression model is:

$$
f_{\vec{w},b}(\vec{x}) = g(\vec{w} \cdot \vec{x} + b)
$$

This output is interpreted as the probability that the input belongs to class one:

$$
f_{\vec{w},b}(\vec{x}) = P(y = 1 \mid \vec{x}; \vec{w}, b)
$$

A common classification threshold is:

$$
f_{\vec{w},b}(\vec{x}) \geq 0.5
$$

If the model output is greater than or equal to the threshold, the prediction is class one:

$$
\hat{y} = 1
$$

If the model output is less than the threshold, the prediction is class zero:

$$
\hat{y} = 0
$$

Because the sigmoid function gives a value of one half when its input is zero:

$$
g(0) = 0.5
$$

The threshold rule can also be written in terms of the linear value:

$$
\vec{w} \cdot \vec{x} + b \geq 0 \quad \Rightarrow \quad \hat{y} = 1
$$

$$
\vec{w} \cdot \vec{x} + b < 0 \quad \Rightarrow \quad \hat{y} = 0
$$

The decision boundary is the set of points where the model is exactly at the threshold:

$$
\vec{w} \cdot \vec{x} + b = 0
$$

For example, if the parameters are:

Then:

$$
z = x_1 + x_2 - 3
$$

The decision boundary is found by setting this equal to zero:

$$
x_1 + x_2 - 3 = 0
$$

Which gives the straight-line boundary:

If the parameters are:

The decision boundary is:

Which gives the circular boundary:

In this example, points outside the circle are predicted as class one:

$$

And points inside the circle are predicted as class zero:

$$
x_1^2 + x_2^2 < 1 \quad \Rightarrow \quad \hat{y} = 0

## 3.2 Cost function for logistic regression

### Why Squared Error Cost Is Not Used for Logistic Regression?

In linear regression, we used the squared error cost function. For one training example, the loss is:

$$
L\left(f_{\vec{w},b}(\vec{x}^{(i)}), y^{(i)}\right) = \frac{1}{2}\left(f_{\vec{w},b}(\vec{x}^{(i)}) - y^{(i)}\right)^2
$$

The full cost function averages this loss over all training examples:

$$
J(\vec{w}, b) = \frac{1}{m}\sum_{i=1}^{m} L\left(f_{\vec{w},b}(\vec{x}^{(i)}), y^{(i)}\right)
$$

Substituting the squared error loss gives:

$$
J(\vec{w}, b) = \frac{1}{m}\sum_{i=1}^{m}\frac{1}{2}\left(f_{\vec{w},b}(\vec{x}^{(i)}) - y^{(i)}\right)^2
$$

For linear regression, the model is linear in the parameters:

$$
f_{\vec{w},b}(\vec{x}) = \vec{w} \cdot \vec{x} + b
$$

Because this model is linear, the squared error cost forms a convex, bowl-shaped function. That means gradient descent has one global minimum to move toward.

Logistic regression uses a sigmoid function instead of a direct linear output:

$$
f_{\vec{w},b}(\vec{x}) = g(z)
$$

$$
z = \vec{w} \cdot \vec{x} + b
$$

$$
g(z) = \frac{1}{1 + e^{-z}}
$$

If we try to reuse the squared error cost with logistic regression, the cost becomes:

$$
J(\vec{w}, b) = \frac{1}{m}\sum_{i=1}^{m}\frac{1}{2}\left(g(\vec{w} \cdot \vec{x}^{(i)} + b) - y^{(i)}\right)^2
$$

This is a problem because the sigmoid function is non-linear. After placing the sigmoid inside the squared error expression, the cost function is usually non-convex. A non-convex cost can have many local minima, so gradient descent may get stuck in a bad local minimum instead of finding the best parameters.

To avoid this, logistic regression uses a different loss function that is designed for probabilities and binary labels.

For one training example, the logistic loss is:

$$
L\left(f_{\vec{w},b}(\vec{x}^{(i)}), y^{(i)}\right) =
\begin{cases}
-\log\left(f_{\vec{w},b}(\vec{x}^{(i)})\right), & \text{if } y^{(i)} = 1 \\
-\log\left(1 - f_{\vec{w},b}(\vec{x}^{(i)})\right), & \text{if } y^{(i)} = 0
\end{cases}
$$

When the true label is one, the loss is small if the model predicts a probability close to one:

$$
y^{(i)} = 1, \quad f_{\vec{w},b}(\vec{x}^{(i)}) \to 1 \quad \Rightarrow \quad L \to 0
$$


But the loss becomes very large if the model predicts a probability close to zero:

$$
y^{(i)} = 1, \quad f_{\vec{w},b}(\vec{x}^{(i)}) \to 0 \quad \Rightarrow \quad L \to \infty
$$

</figure>

When the true label is zero, the loss is small if the model predicts a probability close to zero:

$$
y^{(i)} = 0, \quad f_{\vec{w},b}(\vec{x}^{(i)}) \to 0 \quad \Rightarrow \quad L \to 0
$$

But the loss becomes very large if the model predicts a probability close to one:

$$
y^{(i)} = 0, \quad f_{\vec{w},b}(\vec{x}^{(i)}) \to 1 \quad \Rightarrow \quad L \to \infty
$$

   when the true label is 0
  </figcaption>
</figure>

The two cases can be combined into one equation:

$$
L\left(f_{\vec{w},b}(\vec{x}^{(i)}), y^{(i)}\right) = -y^{(i)}\log\left(f_{\vec{w},b}(\vec{x}^{(i)})\right) - \left(1 - y^{(i)}\right)\log\left(1 - f_{\vec{w},b}(\vec{x}^{(i)})\right)
$$

Therefore, the logistic regression cost function is:

$$
J(\vec{w}, b) = -\frac{1}{m}\sum_{i=1}^{m}\left[y^{(i)}\log\left(f_{\vec{w},b}(\vec{x}^{(i)})\right) + \left(1 - y^{(i)}\right)\log\left(1 - f_{\vec{w},b}(\vec{x}^{(i)})\right)\right]
$$

The key idea is that squared error can make logistic regression difficult to optimize because it creates a non-convex cost surface. Logistic loss fixes this by strongly penalizing confident wrong predictions and producing a cost function that gradient descent can optimize more reliably.

###  Simplified Logistic Loss and Cost

The logistic loss function can first be written as two separate cases, depending on the true label.

$$
L\left(f_{\vec{w},b}\left(\vec{x}^{(i)}\right), y^{(i)}\right) =
\begin{cases}
-\log\left(f_{\vec{w},b}\left(\vec{x}^{(i)}\right)\right), & \text{if } y^{(i)} = 1 \\
-\log\left(1 - f_{\vec{w},b}\left(\vec{x}^{(i)}\right)\right), & \text{if } y^{(i)} = 0
\end{cases}
$$

This means the loss is small when the model predicts close to the correct label, and large when the model is confidently wrong.

When the true label is one, only the first part matters:

$$
y^{(i)} = 1 \quad \Rightarrow \quad L\left(f_{\vec{w},b}\left(\vec{x}^{(i)}\right), y^{(i)}\right) = -\log\left(f_{\vec{w},b}\left(\vec{x}^{(i)}\right)\right)
$$

When the true label is zero, only the second part matters:

$$
y^{(i)} = 0 \quad \Rightarrow \quad L\left(f_{\vec{w},b}\left(\vec{x}^{(i)}\right), y^{(i)}\right) = -\log\left(1 - f_{\vec{w},b}\left(\vec{x}^{(i)}\right)\right)
$$

Because binary labels are either zero or one, both cases can be combined into one compact equation:

$$
L\left(f_{\vec{w},b}\left(\vec{x}^{(i)}\right), y^{(i)}\right) = -y^{(i)}\log\left(f_{\vec{w},b}\left(\vec{x}^{(i)}\right)\right) - \left(1 - y^{(i)}\right)\log\left(1 - f_{\vec{w},b}\left(\vec{x}^{(i)}\right)\right)
$$

If the label is one, then the second term becomes zero:

$$
1 - y^{(i)} = 0
$$

So the loss becomes:

$$
L\left(f_{\vec{w},b}\left(\vec{x}^{(i)}\right), y^{(i)}\right) = -\log\left(f_{\vec{w},b}\left(\vec{x}^{(i)}\right)\right)
$$

If the label is zero, then the first term becomes zero:

$$
y^{(i)} = 0
$$

So the loss becomes:

$$
L\left(f_{\vec{w},b}\left(\vec{x}^{(i)}\right), y^{(i)}\right) = -\log\left(1 - f_{\vec{w},b}\left(\vec{x}^{(i)}\right)\right)
$$

The cost function is the average loss over all training examples:

$$
J\left(\vec{w}, b\right) = \frac{1}{m}\sum_{i=1}^{m} L\left(f_{\vec{w},b}\left(\vec{x}^{(i)}\right), y^{(i)}\right)
$$

Substituting the simplified logistic loss gives the logistic regression cost function:

$$
J\left(\vec{w}, b\right) = -\frac{1}{m}\sum_{i=1}^{m}\left[y^{(i)}\log\left(f_{\vec{w},b}\left(\vec{x}^{(i)}\right)\right) + \left(1 - y^{(i)}\right)\log\left(1 - f_{\vec{w},b}\left(\vec{x}^{(i)}\right)\right)\right]
$$

This cost function is convex for logistic regression, so it has a single global minimum. That makes it much better suited for gradient descent than squared error cost with a sigmoid model.

```python
def compute_cost_logistic(X, y, w, b):
    """
    Computes cost

    Args:
      X (ndarray (m,n)): Data, m examples with n features
      y (ndarray (m,)) : target values
      w (ndarray (n,)) : model parameters  
      b (scalar)       : model parameter
      
    Returns:
      cost (scalar): cost
    """

    m = X.shape[0]
    cost = 0.0
    for i in range(m):
        z_i = np.dot(X[i],w) + b
        f_wb_i = sigmoid(z_i)
        cost +=  -y[i]*np.log(f_wb_i) - (1-y[i])*np.log(1-f_wb_i)
             
    cost = cost / m
    return cost
```

## 3.3 Gradient descent for logistic regression



Gradient descent itself does not change. It is a general optimization algorithm: compute the gradient of the current cost function, then move each parameter in the opposite direction of that gradient.

$$
w_j := w_j - \alpha \frac{\partial J(\mathbf{w}, b)}{\partial w_j}
$$

$$
b := b - \alpha \frac{\partial J(\mathbf{w}, b)}{\partial b}
$$

For logistic regression, the prediction is not a straight line output. It is the sigmoid of the linear model:

$$
f_{\mathbf{w},b}(\mathbf{x}^{(i)}) = g(z^{(i)}) = \frac{1}{1 + e^{-z^{(i)}}}
$$

$$
z^{(i)} = \mathbf{w} \cdot \mathbf{x}^{(i)} + b
$$

The cost function used for logistic regression is the logistic loss cost:

$$
J(\mathbf{w}, b) = -\frac{1}{m}\sum_{i=1}^{m}\left[y^{(i)}\log\left(f_{\mathbf{w},b}(\mathbf{x}^{(i)})\right) + \left(1-y^{(i)}\right)\log\left(1-f_{\mathbf{w},b}(\mathbf{x}^{(i)})\right)\right]
$$

When we take the partial derivatives of this logistic cost, the update equations become:

$$
\frac{\partial J(\mathbf{w}, b)}{\partial w_j} = \frac{1}{m}\sum_{i=1}^{m}\left(f_{\mathbf{w},b}(\mathbf{x}^{(i)}) - y^{(i)}\right)x_j^{(i)}
$$

$$
\frac{\partial J(\mathbf{w}, b)}{\partial b} = \frac{1}{m}\sum_{i=1}^{m}\left(f_{\mathbf{w},b}(\mathbf{x}^{(i)}) - y^{(i)}\right)
$$

So the full gradient descent updates are:

$$
w_j := w_j - \alpha \left[\frac{1}{m}\sum_{i=1}^{m}\left(f_{\mathbf{w},b}(\mathbf{x}^{(i)}) - y^{(i)}\right)x_j^{(i)}\right]
$$

$$
b := b - \alpha \left[\frac{1}{m}\sum_{i=1}^{m}\left(f_{\mathbf{w},b}(\mathbf{x}^{(i)}) - y^{(i)}\right)\right]
$$

These equations look the same as the gradient descent equations for linear regression, but they are not copied from the old squared error cost. They come from differentiating the new logistic loss cost. The reason they simplify to the same form is that the derivative of the sigmoid and the derivative of the logarithms in the logistic loss cancel in a convenient way, leaving the error term:

$$
f_{\mathbf{w},b}(\mathbf{x}^{(i)}) - y^{(i)}
$$

The important difference is the meaning of $f_{\mathbf{w},b}(\mathbf{x})$. In linear regression, it is:

$$
f_{\mathbf{w},b}(\mathbf{x}) = \mathbf{w} \cdot \mathbf{x} + b
$$

In logistic regression, it is:

$$
f_{\mathbf{w},b}(\mathbf{x}) = \frac{1}{1 + e^{-(\mathbf{w}\cdot\mathbf{x}+b)}}
$$

If we used squared error with the sigmoid model, the gradients would include extra sigmoid derivative terms and the cost surface could become non-convex. Logistic loss avoids that problem and gives a convex cost function that gradient descent can optimize reliably.


```python
import copy, math
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from lab_utils_common_Part3 import  dlc, plot_data, plt_tumor_data, sigmoid, compute_cost_logistic
from plt_quad_logistic import plt_quad_logistic, plt_prob
plt.style.use('./deeplearning.mplstyle')
```

```python
X_train = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y_train = np.array([0, 0, 0, 1, 1, 1])
```

```python
fig,ax = plt.subplots(1,1,figsize=(4,4))
plot_data(X_train, y_train, ax)

ax.axis([0, 4, 0, 3.5])
ax.set_ylabel('$x_1$', fontsize=12)
ax.set_xlabel('$x_0$', fontsize=12)
plt.show()
```

```python
def compute_gradient_logistic(X, y, w, b): 
    """
    Computes the gradient for logistic regression 
 
    Args:
      X (ndarray (m,n): Data, m examples with n features
      y (ndarray (m,)): target values
      w (ndarray (n,)): model parameters  
      b (scalar)      : model parameter
    Returns
      dj_dw (ndarray (n,)): The gradient of the cost w.r.t. the parameters w. 
      dj_db (scalar)      : The gradient of the cost w.r.t. the parameter b. 
    """
    m,n = X.shape
    dj_dw = np.zeros((n,))                           #(n,)
    dj_db = 0.

    for i in range(m):
        f_wb_i = sigmoid(np.dot(X[i],w) + b)          #(n,)(n,)=scalar
        err_i  = f_wb_i  - y[i]                       #scalar
        for j in range(n):
            dj_dw[j] = dj_dw[j] + err_i * X[i,j]      #scalar
        dj_db = dj_db + err_i
    dj_dw = dj_dw/m                                   #(n,)
    dj_db = dj_db/m                                   #scalar
        
    return dj_db, dj_dw  
```

```python
X_tmp = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y_tmp = np.array([0, 0, 0, 1, 1, 1])
w_tmp = np.array([2.,3.])
b_tmp = 1.
dj_db_tmp, dj_dw_tmp = compute_gradient_logistic(X_tmp, y_tmp, w_tmp, b_tmp)
print(f"dj_db: {dj_db_tmp}" )
print(f"dj_dw: {dj_dw_tmp.tolist()}" )
```

```python
def gradient_descent(X, y, w_in, b_in, alpha, num_iters): 
    """
    Performs batch gradient descent
    
    Args:
      X (ndarray (m,n)   : Data, m examples with n features
      y (ndarray (m,))   : target values
      w_in (ndarray (n,)): Initial values of model parameters  
      b_in (scalar)      : Initial values of model parameter
      alpha (float)      : Learning rate
      num_iters (scalar) : number of iterations to run gradient descent
      
    Returns:
      w (ndarray (n,))   : Updated values of parameters
      b (scalar)         : Updated value of parameter 
    """
    # An array to store cost J and w's at each iteration primarily for graphing later
    J_history = []
    w = copy.deepcopy(w_in)  #avoid modifying global w within function
    b = b_in
    
    for i in range(num_iters):
        # Calculate the gradient and update the parameters
        dj_db, dj_dw = compute_gradient_logistic(X, y, w, b)   

        # Update Parameters using w, b, alpha and gradient
        w = w - alpha * dj_dw               
        b = b - alpha * dj_db               
      
        # Save cost J at each iteration
        if i<100000:      # prevent resource exhaustion 
            J_history.append( compute_cost_logistic(X, y, w, b) )

        # Print cost every at intervals 10 times or as many iterations if < 10
        if i% math.ceil(num_iters / 10) == 0:
            print(f"Iteration {i:4d}: Cost {J_history[-1]}   ")
        
    return w, b, J_history         #return final w,b and J history for graphing
```

```python
w_tmp  = np.zeros_like(X_train[0])
b_tmp  = 0.
alph = 0.1
iters = 10000

w_out, b_out, _ = gradient_descent(X_train, y_train, w_tmp, b_tmp, alph, iters) 
print(f"\nupdated parameters: w:{w_out}, b:{b_out}")
```

```python
fig,ax = plt.subplots(1,1,figsize=(5,4))
# plot the probability 
plt_prob(ax, w_out, b_out)

# Plot the original data
ax.set_ylabel(r'$x_1$')
ax.set_xlabel(r'$x_0$')   
ax.axis([0, 4, 0, 3.5])
plot_data(X_train,y_train,ax)

# Plot the decision boundary
x0 = -b_out/w_out[0]
x1 = -b_out/w_out[1]
ax.plot([0,x0],[x1,0], c=dlc["dlblue"], lw=1)
plt.show()
```

## 3.4 The problem of overfitting

A model should fit the training data well, but it should also generalize well to new examples. The problem of overfitting and underfitting is about finding the right level of model complexity.

For regression, a model that is too simple may use only a straight line:

$$
f_{\mathbf{w},b}(x) = w_1x + b
$$

A more flexible model may use polynomial features:

$$
f_{\mathbf{w},b}(x) = w_1x + w_2x^2 + b
$$

A model that is too complex may use many high-degree features:

$$
f_{\mathbf{w},b}(x) = w_1x + w_2x^2 + w_3x^3 + w_4x^4 + b
$$

For classification, logistic regression first computes:

$$
z = \mathbf{w} \cdot \mathbf{x} + b
$$

Then it applies the sigmoid function:

$$
f_{\mathbf{w},b}(\mathbf{x}) = g(z) = \frac{1}{1 + e^{-z}}
$$

If the decision boundary is too simple, the model underfits. If the decision boundary is too complicated, the model overfits.

Underfitting is called **high bias** because the model makes a strong wrong assumption about the data. For example, using a straight line when the real pattern is curved forces the model to ignore important structure. The predictions are consistently wrong in the same direction, so the model has a large bias.

$$
\text{Bias}\left(\hat{f}(x)\right) = \mathbb{E}\left[\hat{f}(x)\right] - f(x)
$$

High bias usually means both the training error and cross-validation error are large:

$$
J_{\text{train}} \text{ is high}, \qquad J_{\text{cv}} \text{ is high}
$$

Overfitting is called **high variance** because the model changes too much depending on the exact training examples. It fits small random details and noise in the training set instead of learning the main pattern. Because of this, it performs very well on the training set but poorly on new data.

$$
\text{Var}\left(\hat{f}(x)\right) = \mathbb{E}\left[\left(\hat{f}(x) - \mathbb{E}\left[\hat{f}(x)\right]\right)^2\right]
$$

High variance usually means the training error is small but the cross-validation error is much larger:

$$
J_{\text{train}} \text{ is low}, \qquad J_{\text{cv}} \gg J_{\text{train}}
$$

The bias-variance idea can be summarized by this decomposition:

$$
\mathbb{E}\left[\left(y - \hat{f}(x)\right)^2\right] = \text{Bias}\left(\hat{f}(x)\right)^2 + \text{Var}\left(\hat{f}(x)\right) + \sigma^2
$$

So the goal is not to make the model as simple or as complex as possible. The goal is to choose a model that has low enough bias to capture the real pattern, and low enough variance to generalize well to new examples.

Overfitting happens when a model fits the training set too closely, including noise or random patterns that do not represent the true relationship. This usually gives very low training error but much higher error on new data.

A very flexible model may use many features or high-degree polynomial terms:

$$
f_{\mathbf{w},b}(x) = w_1x + w_2x^2 + w_3x^3 + w_4x^4 + b
$$

Because it has many parameters, it can bend too much and follow the training examples too exactly. There are three common ways to reduce overfitting.

First, collect more training examples. More data makes it harder for the model to fit random noise, because the true pattern becomes clearer as the number of examples increases.

$$
m \uparrow \quad \Rightarrow \quad \text{variance decreases}
$$

Second, select only the most useful features. If the model has too many features and not enough data, it may use irrelevant features to fit noise. Feature selection reduces the number of inputs from all features to a smaller useful subset.

$$
\mathbf{x} = [x_1, x_2, \dots, x_n]
$$

$$
\mathbf{x}_{\text{selected}} = [x_{j_1}, x_{j_2}, \dots, x_{j_k}], \qquad k < n
$$

The disadvantage is that a useful feature might be removed by mistake, so feature selection must be done carefully.

Third, use regularization. Regularization keeps all features, but it discourages the model from using very large parameter values. This makes the model smoother and less likely to overfit.

Regularization is used to reduce **overfitting**. Instead of only minimizing the training error, we add a penalty for large weight values. This encourages the model to learn a smoother, simpler function that generalizes better to new data.

For a polynomial model:

$$
f_{\mathbf{w},b}(x) = w_1x + w_2x^2 + w_3x^3 + w_4x^4 + b
$$

the model can overfit if the weights become very large, because large weights allow the curve to bend too much around the training examples.

The regularized cost function is:

$$
J(\mathbf{w}, b) =
\frac{1}{2m}\sum_{i=1}^{m}\left(f_{\mathbf{w},b}\left(\mathbf{x}^{(i)}\right) - y^{(i)}\right)^2
+
\frac{\lambda}{2m}\sum_{j=1}^{n} w_j^2
$$

The first term measures how well the model fits the training data:

$$
\frac{1}{2m}\sum_{i=1}^{m}\left(f_{\mathbf{w},b}\left(\mathbf{x}^{(i)}\right) - y^{(i)}\right)^2
$$

The second term is the regularization penalty:

$$
\frac{\lambda}{2m}\sum_{j=1}^{n} w_j^2
$$

The parameter $\lambda$ controls the strength of regularization:

- If $\lambda = 0$, there is no penalty on the weights, so the model may overfit.
- If $\lambda$ is small, large weights are only weakly penalized.
- If $\lambda$ is large, large weights become very expensive, so the optimizer is pushed to make the weights smaller.
- If $\lambda$ is extremely large, the weights can become close to zero, and the model may underfit.

This is why increasing $\lambda$ usually decreases the size of the weights. The optimizer is minimizing the whole cost function. When $\lambda$ is large, a large value of $w_j$ adds a big penalty because $w_j^2$ is multiplied by $\lambda$:

$$
\lambda \uparrow \quad \Rightarrow \quad \frac{\lambda}{2m}\sum_{j=1}^{n}w_j^2 \uparrow
$$

To reduce this penalty, gradient descent shrinks the weights:

$$
w_j := w_j - \alpha\left[
\frac{1}{m}\sum_{i=1}^{m}\left(f_{\mathbf{w},b}\left(\mathbf{x}^{(i)}\right) - y^{(i)}\right)x_j^{(i)}
+
\frac{\lambda}{m}w_j
\right]
$$

This can be rewritten as:

$$
w_j := \left(1 - \alpha\frac{\lambda}{m}\right)w_j
-
\alpha\frac{1}{m}\sum_{i=1}^{m}\left(f_{\mathbf{w},b}\left(\mathbf{x}^{(i)}\right) - y^{(i)}\right)x_j^{(i)}
$$

The important part is:

$$
\left(1 - \alpha\frac{\lambda}{m}\right)w_j
$$

When $\lambda$ increases, the multiplier becomes smaller, so each update pulls $w_j$ more strongly toward zero. When $\lambda$ decreases, this shrinking effect becomes weaker, so the model is freer to use larger weights to fit the training data.

Usually, the bias term $b$ is not regularized because it only shifts the prediction up or down; it does not make the model more complex in the same way large feature weights do.

 Why does a larger $\lambda$ make the weights smaller?

Think of regularization as adding a second goal to the model.

$$
\text{total cost} = \text{cost for fitting the data} + \text{cost for large weights}
$$

For one weight, the regularization penalty is proportional to:

$$
\lambda w_j^2
$$

So if $w_j$ is large, then $w_j^2$ is even larger. When $\lambda$ increases, the model pays a much bigger cost for the same weight value.

  </figcaption>

That means the optimizer has a choice:

- Use large weights and fit the training data very closely.
- Use smaller weights and accept a slightly worse fit, but pay a much smaller penalty.

When $\lambda$ is large, the second option often gives a lower total cost.

The gradient descent update also shows the same idea:

$$
w_j := w_j - \alpha\left(\text{data gradient} + \frac{\lambda}{m}w_j\right)
$$

Distribute the update:

$$
w_j := \left(1 - \alpha\frac{\lambda}{m}\right)w_j - \alpha(\text{data gradient})
$$

Focus only on this part:

$$
\left(1 - \alpha\frac{\lambda}{m}\right)w_j
$$

This is the shrinkage part. It multiplies the old weight by a number smaller than $1$.

For example, suppose:

$$
\alpha = 0.1, \qquad m = 10, \qquad w_j = 10
$$

Then the shrink multiplier is:

$$
1 - \alpha\frac{\lambda}{m}
$$

| $\lambda$ | shrink multiplier | weight after shrink part |
|---:|---:|---:|
| $0$ | $1.00$ | $10.0$ |
| $1$ | $0.99$ | $9.9$ |
| $10$ | $0.90$ | $9.0$ |
| $50$ | $0.50$ | $5.0$ |

So the sentence:

> A larger $\lambda$ makes the multiplier smaller, pulling $w_j$ closer to zero more strongly.

means this:

$$
\lambda \uparrow
\quad \Rightarrow \quad
1 - \alpha\frac{\lambda}{m} \downarrow
\quad \Rightarrow \quad
w_j \text{ is multiplied by a smaller number}
\quad \Rightarrow \quad
w_j \text{ moves closer to } 0
$$

When $\lambda$ is smaller, the multiplier is closer to $1$, so the shrinkage is weaker. The weights are not forced to become larger, but the model is allowed to keep or grow larger weights if they help fit the training data.

**Regularized Linear Regression**

Regularized linear regression is linear regression with an extra penalty term added to the cost function. The goal is still to fit the training data well, but now the model is also encouraged to keep the weights small.

For multiple features, the model prediction is:

$$
f_{\mathbf{w},b}\left(\mathbf{x}^{(i)}\right)
=
\mathbf{w}\cdot\mathbf{x}^{(i)} + b
=
\sum_{j=1}^{n} w_j x_j^{(i)} + b
$$

The regularized cost function is:

$$
J(\mathbf{w}, b)
=
\frac{1}{2m}\sum_{i=1}^{m}\left(f_{\mathbf{w},b}\left(\mathbf{x}^{(i)}\right)-y^{(i)}\right)^2
+
\frac{\lambda}{2m}\sum_{j=1}^{n}w_j^2
$$

So the optimization problem is:

$$
\min_{\mathbf{w},b} J(\mathbf{w}, b)
$$

The first part of the cost measures the prediction error:

$$
\frac{1}{2m}\sum_{i=1}^{m}\left(f_{\mathbf{w},b}\left(\mathbf{x}^{(i)}\right)-y^{(i)}\right)^2
$$

The second part is the regularization term:

$$
\frac{\lambda}{2m}\sum_{j=1}^{n}w_j^2
$$

This term penalizes large weights. Notice that regularization is applied to the weights only:

$$
w_1, w_2, \dots, w_n
$$

The bias term is usually not regularized:

$$
b \text{ is not included in } \sum_{j=1}^{n}w_j^2
$$

To use gradient descent, update each weight and the bias simultaneously:

$$
w_j := w_j - \alpha\frac{\partial}{\partial w_j}J(\mathbf{w}, b),
\qquad j = 1,2,\dots,n
$$

$$
b := b - \alpha\frac{\partial}{\partial b}J(\mathbf{w}, b)
$$

The derivative with respect to a weight is:

$$
\frac{\partial}{\partial w_j}J(\mathbf{w}, b)
=
\frac{1}{m}\sum_{i=1}^{m}\left(f_{\mathbf{w},b}\left(\mathbf{x}^{(i)}\right)-y^{(i)}\right)x_j^{(i)}
+
\frac{\lambda}{m}w_j
$$

The derivative with respect to the bias is:

$$
\frac{\partial}{\partial b}J(\mathbf{w}, b)
=
\frac{1}{m}\sum_{i=1}^{m}\left(f_{\mathbf{w},b}\left(\mathbf{x}^{(i)}\right)-y^{(i)}\right)
$$

Therefore, the regularized gradient descent updates are:

$$
w_j := w_j
-
\alpha\left[
\frac{1}{m}\sum_{i=1}^{m}\left(f_{\mathbf{w},b}\left(\mathbf{x}^{(i)}\right)-y^{(i)}\right)x_j^{(i)}
+
\frac{\lambda}{m}w_j
\right],
\qquad j = 1,2,\dots,n
$$

$$
b := b
-
\alpha\left[
\frac{1}{m}\sum_{i=1}^{m}\left(f_{\mathbf{w},b}\left(\mathbf{x}^{(i)}\right)-y^{(i)}\right)
\right]
$$

The extra term in the weight update is:

$$
\frac{\lambda}{m}w_j
$$

This term comes from differentiating the regularization penalty:

$$
\frac{\partial}{\partial w_j}
\left(
\frac{\lambda}{2m}\sum_{k=1}^{n}w_k^2
\right)
=
\frac{\lambda}{2m}\cdot 2w_j
=
\frac{\lambda}{m}w_j
$$

All other weights inside the sum are treated as constants when differentiating with respect to one specific weight:

$$
\frac{\partial}{\partial w_j}w_k^2 = 0
\qquad \text{when } k \ne j
$$

The data-fitting part gives:

$$
\frac{\partial}{\partial w_j}
\left[
\frac{1}{2m}\sum_{i=1}^{m}\left(f_{\mathbf{w},b}\left(\mathbf{x}^{(i)}\right)-y^{(i)}\right)^2
\right]
=
\frac{1}{m}\sum_{i=1}^{m}\left(f_{\mathbf{w},b}\left(\mathbf{x}^{(i)}\right)-y^{(i)}\right)x_j^{(i)}
$$

Combining both parts gives the full derivative for the weight:

$$
\frac{\partial}{\partial w_j}J(\mathbf{w}, b)
=
\frac{1}{m}\sum_{i=1}^{m}\left(f_{\mathbf{w},b}\left(\mathbf{x}^{(i)}\right)-y^{(i)}\right)x_j^{(i)}
+
\frac{\lambda}{m}w_j
$$

A useful way to rewrite the weight update is:

$$
w_j :=
\left(1 - \alpha\frac{\lambda}{m}\right)w_j
-
\alpha\frac{1}{m}\sum_{i=1}^{m}\left(f_{\mathbf{w},b}\left(\mathbf{x}^{(i)}\right)-y^{(i)}\right)x_j^{(i)}
$$

This shows the main effect of regularization: before the normal data-fitting update, the old weight is multiplied by a factor smaller than one:

$$
1 - \alpha\frac{\lambda}{m}
$$

So regularization gradually shrinks the weights toward zero, which helps reduce overfitting and makes the learned model smoother.
Regularized logistic regression is used for classification when the model may overfit, especially when we use many features or high-degree polynomial features. Without regularization, the decision boundary can become too complex and fit the training data too closely.

For logistic regression, we first compute a linear or polynomial expression called `z`:

$$
z
=
w_1x_1 + w_2x_2
+
w_3x_1^2x_2
+
w_4x_1^2x_2^2
+
w_5x_1x_2^2
+
\cdots
+
b
$$

Then we pass `z` through the sigmoid function to get a value between zero and one:

$$
f_{\mathbf{w},b}(\mathbf{x})
=
g(z)
=
\frac{1}{1 + e^{-z}}
$$

This output can be interpreted as the probability that the example belongs to class one:

$$
f_{\mathbf{w},b}(\mathbf{x})
=
P(y = 1 \mid \mathbf{x})
$$

The regularized logistic regression cost function is:

$$
J(\mathbf{w}, b)
=
-\frac{1}{m}\sum_{i=1}^{m}
\left[
y^{(i)}\log\left(f_{\mathbf{w},b}\left(\mathbf{x}^{(i)}\right)\right)
+
\left(1-y^{(i)}\right)\log\left(1-f_{\mathbf{w},b}\left(\mathbf{x}^{(i)}\right)\right)
\right]
+
\frac{\lambda}{2m}\sum_{j=1}^{n}w_j^2
$$

The optimization goal is:

$$
\min_{\mathbf{w},b} J(\mathbf{w}, b)
$$

The first part of the cost is the logistic loss:

$$
-\frac{1}{m}\sum_{i=1}^{m}
\left[
y^{(i)}\log\left(f_{\mathbf{w},b}\left(\mathbf{x}^{(i)}\right)\right)
+
\left(1-y^{(i)}\right)\log\left(1-f_{\mathbf{w},b}\left(\mathbf{x}^{(i)}\right)\right)
\right]
$$

The second part is the regularization penalty:

$$
\frac{\lambda}{2m}\sum_{j=1}^{n}w_j^2
$$

This penalty discourages the model from using very large weights. Large weights can create complicated decision boundaries, while smaller weights usually create smoother decision boundaries.

Gradient descent has the same general form as before. We update all weights and the bias simultaneously:

$$
w_j := w_j - \alpha\frac{\partial}{\partial w_j}J(\mathbf{w}, b),
\qquad j = 1,2,\dots,n
$$

$$
b := b - \alpha\frac{\partial}{\partial b}J(\mathbf{w}, b)
$$

For regularized logistic regression, the derivative with respect to a weight is:

$$
\frac{\partial}{\partial w_j}J(\mathbf{w}, b)
=
\frac{1}{m}\sum_{i=1}^{m}
\left(f_{\mathbf{w},b}\left(\mathbf{x}^{(i)}\right)-y^{(i)}\right)x_j^{(i)}
+
\frac{\lambda}{m}w_j
$$

The derivative with respect to the bias is:

$$
\frac{\partial}{\partial b}J(\mathbf{w}, b)
=
\frac{1}{m}\sum_{i=1}^{m}
\left(f_{\mathbf{w},b}\left(\mathbf{x}^{(i)}\right)-y^{(i)}\right)
$$

So the full gradient descent update for each weight is:

$$
w_j := w_j
-
\alpha
\left[
\frac{1}{m}\sum_{i=1}^{m}
\left(f_{\mathbf{w},b}\left(\mathbf{x}^{(i)}\right)-y^{(i)}\right)x_j^{(i)}
+
\frac{\lambda}{m}w_j
\right],
\qquad j = 1,2,\dots,n
$$

The full gradient descent update for the bias is:

$$
b := b
-
\alpha
\left[
\frac{1}{m}\sum_{i=1}^{m}
\left(f_{\mathbf{w},b}\left(\mathbf{x}^{(i)}\right)-y^{(i)}\right)
\right]
$$

Notice that the bias update does not include the regularization term. The regularization penalty is applied only to the weights:

$$
w_1,w_2,\dots,w_n
$$

and not to the bias:

$$
b
$$

The weight update can also be rewritten as:

$$
w_j :=
\left(1 - \alpha\frac{\lambda}{m}\right)w_j
-
\alpha\frac{1}{m}\sum_{i=1}^{m}
\left(f_{\mathbf{w},b}\left(\mathbf{x}^{(i)}\right)-y^{(i)}\right)x_j^{(i)}
$$

This form shows the effect of regularization clearly. The factor multiplying the old weight is:

$$
1 - \alpha\frac{\lambda}{m}
$$

When regularization is stronger, the weights are pulled closer to zero, which helps prevent the decision boundary from becoming too complex.


### Regularized Regression Implementation

The following code cells implement the regularized linear and logistic regression cost and gradient functions.


```python
import numpy as np
import matplotlib.pyplot as plt
from plt_overfit import overfit_example, output
from lab_utils_common_Part3 import sigmoid
np.set_printoptions(precision=8)
```

```python
def compute_cost_linear_reg(X, y, w, b, lambda_ = 1):
    """
    Computes the cost over all examples
    Args:
      X (ndarray (m,n): Data, m examples with n features
      y (ndarray (m,)): target values
      w (ndarray (n,)): model parameters  
      b (scalar)      : model parameter
      lambda_ (scalar): Controls amount of regularization
    Returns:
      total_cost (scalar):  cost 
    """

    m  = X.shape[0]
    n  = len(w)
    cost = 0.
    for i in range(m):
        f_wb_i = np.dot(X[i], w) + b                                   #(n,)(n,)=scalar, see np.dot
        cost = cost + (f_wb_i - y[i])**2                               #scalar             
    cost = cost / (2 * m)                                              #scalar  
 
    reg_cost = 0
    for j in range(n):
        reg_cost += (w[j]**2)                                          #scalar
    reg_cost = (lambda_/(2*m)) * reg_cost                              #scalar
    
    total_cost = cost + reg_cost                                       #scalar
    return total_cost      
```

```python
def compute_cost_logistic_reg(X, y, w, b, lambda_ = 1):
    """
    Computes the cost over all examples
    Args:
    Args:
      X (ndarray (m,n): Data, m examples with n features
      y (ndarray (m,)): target values
      w (ndarray (n,)): model parameters  
      b (scalar)      : model parameter
      lambda_ (scalar): Controls amount of regularization
    Returns:
      total_cost (scalar):  cost 
    """

    m,n  = X.shape
    cost = 0.
    for i in range(m):
        z_i = np.dot(X[i], w) + b                                      #(n,)(n,)=scalar, see np.dot
        f_wb_i = sigmoid(z_i)                                          #scalar
        cost +=  -y[i]*np.log(f_wb_i) - (1-y[i])*np.log(1-f_wb_i)      #scalar
             
    cost = cost/m                                                      #scalar

    reg_cost = 0
    for j in range(n):
        reg_cost += (w[j]**2)                                          #scalar
    reg_cost = (lambda_/(2*m)) * reg_cost                              #scalar
    
    total_cost = cost + reg_cost                                       #scalar
    return total_cost        
```

```python
def compute_gradient_linear_reg(X, y, w, b, lambda_): 
    """
    Computes the gradient for linear regression 
    Args:
      X (ndarray (m,n): Data, m examples with n features
      y (ndarray (m,)): target values
      w (ndarray (n,)): model parameters  
      b (scalar)      : model parameter
      lambda_ (scalar): Controls amount of regularization
      
    Returns:
      dj_dw (ndarray (n,)): The gradient of the cost w.r.t. the parameters w. 
      dj_db (scalar):       The gradient of the cost w.r.t. the parameter b. 
    """
    m,n = X.shape           #(number of examples, number of features)
    dj_dw = np.zeros((n,))
    dj_db = 0.

    for i in range(m):                             
        err = (np.dot(X[i], w) + b) - y[i]                 
        for j in range(n):                         
            dj_dw[j] = dj_dw[j] + err * X[i, j]               
        dj_db = dj_db + err                        
    dj_dw = dj_dw / m                                
    dj_db = dj_db / m   
    
    for j in range(n):
        dj_dw[j] = dj_dw[j] + (lambda_/m) * w[j]

    return dj_db, dj_dw
```

```python
def compute_gradient_logistic_reg(X, y, w, b, lambda_): 
    """
    Computes the gradient for linear regression 
 
    Args:
      X (ndarray (m,n): Data, m examples with n features
      y (ndarray (m,)): target values
      w (ndarray (n,)): model parameters  
      b (scalar)      : model parameter
      lambda_ (scalar): Controls amount of regularization
    Returns
      dj_dw (ndarray Shape (n,)): The gradient of the cost w.r.t. the parameters w. 
      dj_db (scalar)            : The gradient of the cost w.r.t. the parameter b. 
    """
    m,n = X.shape
    dj_dw = np.zeros((n,))                            #(n,)
    dj_db = 0.0                                       #scalar

    for i in range(m):
        f_wb_i = sigmoid(np.dot(X[i],w) + b)          #(n,)(n,)=scalar
        err_i  = f_wb_i  - y[i]                       #scalar
        for j in range(n):
            dj_dw[j] = dj_dw[j] + err_i * X[i,j]      #scalar
        dj_db = dj_db + err_i
    dj_dw = dj_dw/m                                   #(n,)
    dj_db = dj_db/m                                   #scalar

    for j in range(n):
        dj_dw[j] = dj_dw[j] + (lambda_/m) * w[j]

    return dj_db, dj_dw  
```
